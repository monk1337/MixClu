```python
from mixclu import *

from IPython.display import display
pd.options.display.max_columns = None

df     = pd.read_csv('./datasets/church.txt')
labels = df.pop('labels')
```

```python
df.head(5)
```

```python
id_col, cat_columns, con_col = get_types(df)
con_col.extend(id_col)
scores                   = pd.DataFrame()
```

```python
"""Model : 1"""

# 1. Cluster based on continuous data only

df_c       = df[con_col]
result_1   = kmeans_model(df_c, 
                         4, 
                         random_state = 0, 
                         max_iter     = 300, 
                         verbose      = 0, 
                         df_output    = False)

scores = evaluate_clusters(scores,  list(result_1['clusters']),
                               labels, name='kmeans_cont', 
                               X=None)
scores
```

```python
"""Model : 2"""

# base kmeans one hot model , scale : None

result_2 = kmeans_onehot_mix(df, 
               cat_columns, 
               4,
               random_state = 0,
               max_iter     = 300, 
               verbose      = 0, 
               df_output    = False,
               scale        = None)

scores = evaluate_clusters(scores,  list(result_2['clusters']),
                               labels, name='kmeans_onehot', 
                               X=None)
scores
```

```python
"""Model : 5"""

# using grower distance as embeddings

result_5, emd = kmoid_grower(df, 
                       cat_columns, 
                       no_of_clusters = 4, 
                       max_iter     = 300, 
                       random_state = 4, 
                       df_output    = False)


scores = evaluate_clusters(scores,  list(result_5['clusters']),
                               labels, name='grower_distance', 
                               X=emd)
scores
```

```python
"""Model : 6"""

# using famd as embeddings -> model


model_6, famd_emd = famd_kemans(df, 
                        cat_columns, 
                        clusters = 4, 
                        random_state = 32, 
                        max_iter = 300, 
                        verbose = 0, 
                        df_output = False)

scores = evaluate_clusters(scores,  list(model_6['clusters']),
                               labels, name='famd_kmodel', 
                               X=famd_emd)
scores
```

```python
"""Model : 7"""

# using umap as embeddings -> model


model_7, umap_emd = umap_embd_model(df, 
                    cat_columns, 
                    no_of_clusters = 4, 
                    random_state = 32, 
                    max_iter     = 300, 
                    verbose      = 0, 
                    df_output    = False)

scores = evaluate_clusters(scores,  list(model_7['clusters']),
                               labels, name='umap_kmodel', 
                               X=umap_emd)
scores
```

```python
"""Model : 8"""

# (Mirkin, 2005, pp. 85–91).
# each dummy variable is standardized by shifting the origin to the mean and dividing it by the standard deviation
# using zscore as embeddings -> model

z_score, result_8 = Mirkin_model(df, 
                                 cat_columns, 
                                 no_of_clusters = 4, 
                                 random_state = 32, 
                                 max_iter = 300, 
                                 verbose =0, 
                                 df_output = False)


scores = evaluate_clusters(scores,  list(result_8['clusters']),
                               labels, name='zscore_kmodel', 
                               X=z_score)
scores
```

```python
"""Model : 9"""

# zscore => then kprototype model

zscore_data, result_9 = k_prototype_z_model(df, 
                        cat_columns, 
                        num_of_clusters = 4, 
                        init_method = 'Cao', 
                        random_state= 32, 
                        max_iter    = 300, 
                        n_jobs      = 1, 
                        verbose     = 0, 
                        df_output   = False, 
                        n_init      = 10)


scores = evaluate_clusters(scores,  list(result_9['clusters']),
                               labels, name='zscore_kprototype', 
                               X=None)
scores
```

```python
"""Model : 10"""

# converting continuous data into kbins and using as categorical features


kbins_model = kmodes_cont(df, 
                            bin_con_columns  = con_col,
                            n_of_clusters    = 4,
                            bin_bins         = 5, 
                            bin_drop_cols    = True, 
                            bin_encode       = False, 
                            mode_init_method = 'Cao', 
                            mode_n_init      = 10, 
                            mode_verbose     = 0, 
                            mode_random_state= 32, 
                            mode_n_jobs      = 1, 
                            mode_cat_dissim  = None, 
                            mode_df_output   = None)

scores = evaluate_clusters(scores,  list(kbins_model['clusters']),
                               labels, name='kbins_model', 
                               X=None)
scores
```

```python
"""Model : 10"""

# converting continuous data into kbins => umap => model


umap_kbins_emd, kbins_umap_model = categorical_embedding_model(df, 
                                                            bin_con_columns = con_col, 
                                                            no_of_clusters = 4,
                                                            bin_bins = 5,
                                                            bin_drop_cols = False,
                                                            bin_encode = False,
                                                            n_neighbors=10, 
                                                            min_dist=0.1, 
                                                            random_state = 32, 
                                                            max_iter = 300, 
                                                            verbose = 0, 
                                                            df_output = False)

scores = evaluate_clusters(scores,  list(kbins_umap_model['clusters']),
                               labels, name='kbins_umap_model', 
                               X=umap_kbins_emd)
scores
```

```python
"""Model : 11"""

# converting data into kbins = > then graph and then community detection

graph_model = graph_clustering(df, con_col,
                     bins = 5, 
                     bin_drop_columns = False, 
                     bin_encode = False, 
                     df_output = False)


scores = evaluate_clusters(scores,  list(graph_model['clusters']),
                               labels, name='graph_model', 
                               X=None)
scores
```
